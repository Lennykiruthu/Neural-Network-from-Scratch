```python
import matplotlib.pyplot as plt
import numpy as np

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
```

```python
# Load images as a 3D tensor (rows, columns, depth)
(train_images, train_labels), _ = mnist.load_data()

print(f'Training images first row \n{train_images[0]}')
print(f'\nTraining images shape {train_images.shape}')

# Display images before preprocessing as plt.imshow() expects a 3D tensor but
# modelling expects a 2D shapes
plt.figure(figsize=(6,6))
for i in range(9):
  plt.subplot(1,9,i+1)
  plt.imshow(train_images[i], cmap='gray')
  plt.ylabel(f'{train_labels[i]}')
  plt.axis('off')

print('\n Printing the first 9 images for reference')
plt.tight_layout()
plt.show()

# Preprocessing but reducing dimensions, converting to float32 and standardizing
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

print(f'Training images first row after preprocessing \n{train_images[0]}')
print(f'\nTraining images shape {train_images.shape}')
print(f'\nTraining labels shape {train_labels.shape}')
```

```python
train_images_with_noise = np.concatenate([train_images,
                                         np.random.random((len(train_images), 784))]
                                         , axis=1)
train_images_with_zeros = np.concatenate([train_images,
                                         np.zeros((len(train_images), 784))]
                                         , axis=1)

print(f'Training image with noise first row: {train_images_with_noise[0]}')
print(f'Training image with noise shape: {train_images_with_noise.shape}')
print(f'Training image with zeros first row: {train_images_with_zeros[0]}')
print(f'Training image with zeros shape: {train_images_with_zeros.shape}')
```

```python
# Define function for model building
def build_model():
  model = keras.Sequential([
      layers.Dense(512, activation='relu'),
      layers.Dense(10, activation='softmax')
  ])
  model.compile(optimizer='rmsprop',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
  return model

# Initialize noise model
noise_model = build_model()
noise_history = noise_model.fit(train_images_with_noise, train_labels,
                                epochs=10, batch_size=128,
                                validation_split=0.2)

# Initialize zeros mode
zeros_model = build_model()
zeros_history = zeros_model.fit(train_images_with_zeros, train_labels,
                                epochs=10, batch_size=128,
                                validation_split=0.2)
```

```python
# Retrive val accuracy scores and epochs count
noise_val_accuracy = noise_history.history['val_accuracy']
zeros_val_accuracy = zeros_history.history['val_accuracy']
epochs = range(1,11)

plt.plot(epochs, noise_val_accuracy, 'b', label='Noise Validation Accuracy')
plt.plot(epochs, zeros_val_accuracy, 'b--', label='Zeros Validation Accuracy')
plt.title('Validation Accuracies with Noise and Zeros')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```
